**Importar bibliotecas**

```python
import nltk
```

```python
nltk.download('gutenberg')
```

```python
nltk.download('punkt')
```

```python
biblioteca = nltk.corpus.gutenberg.fileids()
```

**Quantidade de palavras do livro Moby Dick**

```python
mobyDick = nltk.corpus.gutenberg.words('melville-moby_dick.txt')
```

```python
mobyDick
```

```python
len(mobyDick)
```

```python
len(set(mobyDick))
```

```python
(len(set(mobyDick))/len(mobyDick)) * 100
```

Capturar sentenças

```python
sents = nltk.corpus.gutenberg.sents("melville-moby_dick.txt")
sents
```

```python
len(sents)
```

```python
alice = nltk.corpus.gutenberg.words('carroll-alice.txt')
len(alice)
```

```python
sents_in_alice = nltk.corpus.gutenberg.sents('carroll-alice.txt')
len(sents_in_alice)
```

Operações com documentos

```python
for documents in biblioteca:
    livro = documents
    palavras = len(nltk.corpus.gutenberg.words(documents))
    print(livro, palavras)
```

Frequencia de palavras

```python
freq_alice = nltk.FreqDist(alice)
```

```python
freq_alice['Rabbit'] + freq_alice['rabbit']
```

```python
freq_alice.most_common(50)
```

```python
freq_baleia = nltk.FreqDist(mobyDick)
freq_baleia
```

```python
freq_baleia.most_common(50)
```
