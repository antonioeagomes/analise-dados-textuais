```python
import nltk
```

```python
import re
```

```python
caesar = nltk.corpus.gutenberg.words('shakespeare-caesar.txt')
```

**Para o corpus “shakespeare-caesar.txt”, usando expressões regulares responda os seguintes itens:**

**a) Quantidades de palavras que terminam com “r”**

```python
def search_regex(expressao):
    lista_palavras = []
    for word in caesar:
        if re.search(expressao, word):
            lista_palavras.append(word)
    return set(lista_palavras)

print(len(search_regex('r$')))
```

**b) Quantidade de palavras com 5 letras;**

```python
print(len(search_regex('^([a-zA-Z]{5})$')))
```

**c) Quantidade de vezes que “err” ocorre no corpus**

```python
print(search_regex('(err)'))

lista_palavras = []
for word in caesar:
    if re.search('(err)', word):
        lista_palavras.append(word)
print(lista_palavras)
print(len(lista_palavras))
```

**d) Quantidade de vezes que “are” ocorre no corpus para as palavras com 5 ou mais caracteres.**

```python
are = search_regex('are')
print(re.findall('{5}', are))
```

**2) Em relação do corpus “shakespeare-hamlet.txt”, faça as seguintes atividades:**

**a) Normalize o corpus (Retire os números e deixe todas as palavras minúsculas)**

```python
hamlet = nltk.corpus.gutenberg.words('shakespeare-hamlet.txt')
hamlet_norm = [word.lower() for word in hamlet if word.isalpha()]
hamlet_norm
```

**b) Aplique o lematizador em todas as palavras do corpus;**

```python
lema = nltk.WordNetLemmatizer()
lematizadas = []
for palavra in hamlet_norm:
    lematizadas.append(lema.lemmatize(palavra))
lematizadas
```

**c) Aplique o tokenizador em todas as sentenças do corpus;**

```python
hamlet_sents = nltk.corpus.gutenberg.sents('shakespeare-hamlet.txt')
hamlet_sents
tokens = []
for phrase in hamlet_sents:
    #tokens.append(' '.join(phrase))
    tokens.append(nltk.word_tokenize(' '.join(phrase)))
tokens
```

**d) Aplique os pos tagger e responda a quantidade de adjetivos existem no corpus; Dica a tag de adjetivo é ​ “JJ”**

```python
hamlet_tags = nltk.pos_tag(hamlet_norm)
hamlet_tags
```

```python
adjectives = [word for word in hamlet_tags if word[1] == 'JJ']
print(len(adjectives))
print(adjectives)
```
