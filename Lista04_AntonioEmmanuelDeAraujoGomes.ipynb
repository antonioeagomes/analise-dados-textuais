```python
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

**1) Retire as stopwords dos seguintes textos (Esses textos pertencem
ao corpus gutemberg):**

**a) shakespeare-caesar.txt**

```python
caesar = nltk.corpus.gutenberg.words('shakespeare-caesar.txt')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
processadas_caesar = []
normalizada = [word.lower() for word in caesar if word.isalpha()]
for word in normalizada:    
    if word not in stopwords:
        processadas_caesar.append(word)
print(processadas_caesar)
```

```python
print(len(processadas_caesar))
```

**b) shakespeare-hamlet.txt**

```python
hamlet = nltk.corpus.gutenberg.words('shakespeare-hamlet.txt')
```

```python
palavras_relevantes_hamlet = [word.lower() for word in hamlet if word.isalpha() and word not in stopwords]
```

```python
print(len(palavras_relevantes_hamlet))
```

**3) Encontre as entidades nomeada presentes para o seguintes textos:**

**a) apoloxi.txt**

```python
token_punkt = tokenize.WordPunctTokenizer()
apoloXI = open('apoloxi.txt').read()
apolo_tag = nltk.pos_tag(token_punkt.tokenize(apoloXI))
apolo_tag
```

**b) french-revolution.txt**

```python
french_revs = open('french-revolution.txt').read()
french_tag = nltk.pos_tag(token_punkt.tokenize(french_revs))
french_tag
```

**4) Qual é a quantidade de entidades "GPE" presentes em cada um dos textos?**

```python
french_ch = nltk.ne_chunk(french_tag)
len(french_ch)
```

```python
french_ch[:10]
```

```python
apoloXi_ch = nltk.ne_chunk(apolo_tag)
len(apoloXi_ch)
apoloXI_gpe = []
for entity in apoloXi_ch:
   if hasattr(entity, 'label'):
    if (entity.label() == 'GPE'):
        apoloXI_gpe.append(' '.join(c[0] for c in entity.leaves()))
len(apoloXI_gpe)
```

```python
apoloXI_gpe
```

```python
french_gpe = []
for entity in french_ch:
   if hasattr(entity, 'label'):
    if (entity.label() == 'GPE'):
        french_gpe.append(' '.join(c[0] for c in entity.leaves()))
len(french_gpe)
```
